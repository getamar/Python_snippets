```python
x = np.linspace(0,10,1000)
fig = plt.figure()
# plt.plot(x, func(x),'-'), third parameter is what symbol to show in the graph
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
plt.plot(x,np.cos(x) + np.sin(x),'.')
plt.show()
```

```python
# Basic error bars
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * (np.random.randn(50))
plt.errorbar(x,y,yerr = dy, fmt ='.r', ecolor = 'lightgray', elinewidth = 3, capsize = 0)
# plt.errorbar(x,y,yerr = dy, fmt ='.r')
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3,2, 1000)
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins = 40)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
# kwargs is passing the function in the python
# alpha is for transparency
```

```python
# two dimensional histogram
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,10000).T
```

```python
plt.hist2d(x,y,bins = 30, cmap ='pink')
cb = plt.colorbar()
cb.set_label('counts in bin')
```
