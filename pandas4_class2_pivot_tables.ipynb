```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.shape
```

```python
titanic.groupby('alive').mean()
```

```python
titanic.groupby(['sex','class'])['survived'].aggregate('mean').unstack()
```

```python
titanic.pivot_table('survived', index ='sex', columns = 'class')
```

```python
titanic.pivot_table('survived', index ='class', columns = 'sex')
```

```python
titanic.pivot_table('survived', index ='class', columns = 'sex', aggfunc = 'count')
```

```python
titanic.groupby(['sex','alive']).count()
```

```python
age = pd.cut(titanic['age'],[0,18,80])
titanic.pivot_table('survived',['sex',age],'class')
```

```python
age = pd.cut(titanic['age'],np.linspace(0,100,9))
```

```python
titanic.pivot_table('survived',['sex',age],'class')
```

```python
age = pd.cut(titanic['age'],[0,18,80])
fare = pd.qcut(titanic['fare'],2)
titanic.pivot_table('survived',['sex',age],[fare,'class'])
```

```python
titanic.pivot_table(index='sex',columns='class',
                    aggfunc={'survived':sum, 'fare':'mean'})
```
